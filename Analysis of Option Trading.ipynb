```python

def describe_data(stocks_df):
  """Function to describe data in terms of unique va"""
  stock_col=[]
  dtypes=[]
  unique_val=[]
  count=[]
  missing_val=[]
  missing_percentage=[]

  for x in stocks_df.columns:
    stock_col.append(x)
    dtypes.append(stocks_df[x].dtype)
    unique_val.append(len(stocks_df[x].unique()))
    count.append(len(stocks_df[x]))
    missing_val.append(stocks_df[x].isna().sum())
    missing_percentage.append(round((stocks_df[x].isna().sum() / len(stocks_df[x])) * 100, 2))
  
  output=pd.DataFrame({
      'Col_Name': stock_col, 
        'dtype': dtypes,
        'count': count,
        'unique': unique_val,
        'missing': missing_val, 
        'missing_%': missing_percentage
  })
  return output

describe_data(stocks_df)
```

Using the .isna().sum() we can find out whether the dataset contains null value or not. By looking at the table, we saw that "*" contains the null value. We also found out that "*" has missing value which must be removed.

```python
from sklearn.model_selection import train_test_split
train_data = stocks_df[stocks_df['date'] < '2021-07-30']  # Training data from July 1 to July 29
test_data = stocks_df[stocks_df['date'] == '2021-07-30']  # Testing data for July 30

train_data

```

```python
test_data.columns
```

```python
x_val=['mean price', 'iv',
       'volume', 'stock price for iv', 'delta', 'vega', 'gamma', 'theta',
       'rho']
y_val=['adjusted close']
X_train = train_data[x_val]
Y_train = train_data[y_val]
X_test = test_data[x_val]
Y_test = test_data[y_val]

from sklearn.linear_model import LinearRegression

lm=LinearRegression()
lm.fit(X_train,Y_train)

```

**REGRESSION Evaluation Metrics**

MAE : Mean Absolute Error

MSE : Mean Squared Error

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(Y_test,prediction))
print("MSE: ",metrics.mean_squared_error(Y_test,prediction))
print("RMSE: ",np.sqrt(metrics.mean_squared_error(Y_test,prediction)))
```
